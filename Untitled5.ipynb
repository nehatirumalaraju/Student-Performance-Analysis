```python
# STEP 1: IMPORTING DATASET + LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("StudentPerformanceFactors.csv")

print("✅ Dataset Loaded Successfully")
print(df.head())


# STEP 2: DATA EXPLORATION
print("\n--- DATA EXPLORATION ---")
print("Shape of Dataset:", df.shape)
print("\nDataset Info:")
print(df.info())
print("\nFirst 5 rows:")
print(df.head())
print("\nSummary Statistics:")
print(df.describe())
print("\nMissing Values:")
print(df.isnull().sum())

# STEP 3: DATA CLEANING
print("\n--- DATA CLEANING ---")
df = df.dropna() 
df = df.drop_duplicates()

# Feature Engineering - Use Exam_Score as KPI
df['Result'] = df['Exam_Score'].apply(lambda x: 'Pass' if x >= 50 else 'Fail') # 50% cutoff
df['Performance_Level'] = pd.cut(df['Exam_Score'], bins=[0,40,60,80,100], labels=['Low','Average','Good','Excellent'])

print("New columns added: Result, Performance_Level")


# STEP 4: DATA GROUPING
print("\n--- DATA GROUPING ---")
# 1. Performance by Gender
gender_perf = df.groupby('Gender').agg({
    'Exam_Score': 'mean',
    'Result': lambda x: (x=='Pass').mean()*100
}).round(2)
gender_perf.rename(columns={'Exam_Score':'Avg_Exam_Score','Result':'Pass_Rate_%'}, inplace=True)
print(gender_perf)

# 2. Impact of Tutoring
tutor_perf = df.groupby('Tutoring_Sessions')['Exam_Score'].mean().round(2)
print("\nAvg Score by Tutoring Sessions:")
print(tutor_perf)

# 3. Top 10 Students
toppers = df.nlargest(10, 'Exam_Score')[['Gender','Hours_Studied','Attendance','Exam_Score']]

# Save automated report
import os
os.makedirs('Student_Reports', exist_ok=True)  # New folder name with underscore

with pd.ExcelWriter('Student_Reports/Automated_Student_Report.xlsx', engine='openpyxl') as writer:
    df.to_excel(writer, sheet_name='Cleaned_Data', index=False)
    gender_perf.to_excel(writer, sheet_name='Gender_Analysis')
    tutor_perf.to_excel(writer, sheet_name='Tutoring_Analysis')
    toppers.to_excel(writer, sheet_name='Top_10')
    
print("✅ Excel Report Saved Successfully in 'Student_Reports' folder")

# STEP 5: DATA VISUALISATION
print("\n--- DATA VISUALISATION ---")
plt.style.use('seaborn-v0_8-whitegrid')

# Chart 1: Exam Score Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Exam_Score'], bins=20, kde=True)
plt.title("Exam Score Distribution")
plt.show()

# Chart 2: Hours Studied vs Exam Score
plt.figure(figsize=(8,5))
sns.scatterplot(x='Hours_Studied', y='Exam_Score', hue='Gender', data=df)
plt.title("Impact of Hours Studied on Exam Score")
plt.show()

# Chart 3: Pass vs Fail %
plt.figure(figsize=(6,6))
df['Result'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Pass vs Fail Percentage")
plt.ylabel('')
plt.show()
```
